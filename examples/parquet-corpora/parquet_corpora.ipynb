```python
from languagechange.corpora import SprakBankenCorpus, ParquetCorpus
from languagechange.search import SearchTerm
```

<p>Load a corpus in XML format from Språkbanken and use it in a corpus object and cast to a parquet file:</p>

```python
!wget https://spraakbanken.gu.se/resurser/meningsmangder/svt-2004.xml.bz2
```

```python
xml_corpus = SprakBankenCorpus("svt-2004.xml.bz2")
xml_corpus.cast_to_parquet()
```

<p>Load the parquet corpus from the newly created file:</p>

```python
parquet_corpus = ParquetCorpus("svt-2004.parquet")
```

<p>Search for some terms (lemmas 'kvinna' and 'man'):</p>

```python
from time import time

search_terms = [SearchTerm(lemma = "kvinna"), SearchTerm(lemma="man")]

t1 = time()
xml_results = xml_corpus.search(search_terms)
t2 = time()
print(f"Search took {t2 - t1} s.")
parquet_results = parquet_corpus.search(search_terms)
t3 = time()
print(f"Search took {t3 - t2} s.")
```

<p>Look at some of the results:</p>

```python
for u in parquet_results["lemma=kvinna"][:10]:
    print(u)
```
